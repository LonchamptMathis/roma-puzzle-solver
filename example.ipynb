# Roma Solver - Experimentation

## Imports

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from roma import RomaPuzzle
```

## Roma Puzzle Instance

```python
# Example usage
n = 4
roma_cell = (1, 1)
arrows = {(0, 0): '↓', (0, 1): '→', (1, 3): '↑', (3, 1): '←', (3, 2): '↑', (3, 3): '↑'}
boxes = {
    (0, 0): 1, (1, 0): 2, (2, 0): 2, (3, 0): 2,
    (0, 1): 3, (0, 2): 3, (1, 2): 3, (2, 2): 3,
    (0, 3): 4, (1, 3): 4, (2, 3): 5, (3, 3): 5,
    (3, 2): 6, (3, 1): 6, (2, 1): 6, (1, 1): 7
}

# Create the puzzle
puzzle = RomaPuzzle(n, roma_cell, arrows, boxes)
puzzle.display_board()
```

```python
# Retrieve adjacency matrix
adj_matrix = puzzle.adjacency_matrix()
print(np.array(adj_matrix))
```

```python
# Display graph
puzzle.display_graph()
```
